## Heart Disease Analysis

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

[Seaborn](https://seaborn.pydata.org/index.html) is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
data = pd.read_csv('heart.csv', 
                   dtype={'sex': 'object',
                          'cp': 'object', 
                          'fbs': 'object', 
                          'restecg': 'object', 
                          'exang': 'object', 
                          'slope' : 'object', 
                          'thal':'object'})
```

```python
data.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),annot=True,fmt='.1f')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
data.age.value_counts()[:10]
```

```python
sns.barplot(x = data.age.value_counts()[:10].index,
            y = data.age.value_counts()[:10].values)
plt.xlabel('Age')
plt.ylabel('Age Counter')
plt.title('Age Analysis System')
plt.show()

```

```python
sns.relplot(x="trestbps", y="age",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=data)
```

```python
data.plot(kind="scatter", x="age", y="chol", alpha= 0.5, color="brown")
plt.xlabel("age")
plt.ylabel("chol")
plt.title("age and chol with Scatter Plot")
plt.show()
```
